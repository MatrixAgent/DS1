```python
import pandas as pd
authors = pd.DataFrame({'author_id': [1, 2, 3], 'author_name': ['Тургенев', 'Чехов', 'Островский']})
```

```python
authors
```

```python
books = pd.DataFrame({'author_id': [1, 1, 1, 2, 2, 3, 3], 'book_title': ['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'], 'price': [450, 300, 350, 500, 450, 370, 290]})
```

```python
books
```

```python
authors_price = authors.merge(books, how='inner')
```

```python
authors_price
```

```python
top5 = authors_price.sort_values('price', ascending=False).head()
```

```python
top5
```

```python
authors_price.groupby('author_name')['price'].agg(min_price='min', max_price='max', mean_price='mean')
```

```python
authors_price['cover'] = ['твердая', 'мягкая', 'мягкая', 'твердая', 'твердая', 'мягкая', 'мягкая']
authors_price
```

```python
book_info = authors_price.pivot_table(index='author_name', columns='cover', values='price', aggfunc='sum')
book_info
```

```python
book_info.fillna(0, inplace=True)
book_info
```

```python
book_info.to_pickle('book_info.pkl')
```
